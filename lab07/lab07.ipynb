```julia
using Roots  
using Plots
using ForwardDiff
```

```julia
function f1(x)
    return sin(x) - x/2
end

function f2(x)
    return 2*x - exp(-x)
end

function f3(x)
    return x*exp(x)
end

function f3(x)
    return x*exp(-x)
end

function f4(x)
    return cos(x)-x
end

function f5(x)
    return ( x + 3 ) * ( x - 1 )^2
end

function f6(x)
    return (x-1)^7
end
```

```julia
using Pkg
Pkg.add("DataFrames")
```

```julia
using DataFrames
df = DataFrame(Function = String[],Type = String[], X_0 = Float64[], Is_zero = Bool[])
D(f) = x->ForwardDiff.derivative(f, float(x))

x = find_zero(f1, (-1, 1), verbose=true)
push!(df,("f1","przedzial",x,iszero(f1(x))))

x = find_zero((f1, D(f1)),0, Roots.Newton(),verbose=true)
push!(df,("f1","pochodna",x,iszero(f1(x))))

x = find_zero(f1,0,  Order1(),verbose=true)
push!(df,("f1","sieczne",x,iszero(f1(x))))

x = find_zero(f2, (-1, 1), verbose=true)
push!(df,("f2","przedzial",x,iszero(f2(x))))

x = find_zero((f2, D(f2)),0, Roots.Newton(),verbose=true)
push!(df,("f2","pochodna",x,iszero(f2(x))))

x = find_zero(f2,0,  Order1(),verbose=true)
push!(df,("f2","sieczne",x,iszero(f2(x))))

x = find_zero(f3, (-1, 1), verbose=true)
push!(df,("f3","przedzial",x,iszero(f3(x))))

x = find_zero((f3, D(f3)),0, Roots.Newton(),verbose=true)
push!(df,("f3","pochodna",x,iszero(f3(x))))

x = find_zero(f3,0,  Order1(),verbose=true)
push!(df,("f3","sieczne",x,iszero(f3(x))))

x = find_zero(f4, 0, verbose=true)
push!(df,("f4","przedzial",x,iszero(f4(x))))

x = find_zero((f4, D(f4)),0, Roots.Newton(),verbose=true)
push!(df,("f4","pochodna",x,iszero(f4(x))))

x = find_zero(f4,0,  Order1(),verbose=true)
push!(df,("f4","sieczne",x,iszero(f4(x))))

x = find_zero(f5, 0, verbose=true)
push!(df,("f5","przedzial",x,iszero(f5(x))))

x = find_zero((f5, D(f5)),0, Roots.Newton(),verbose=true)
push!(df,("f5","pochodna",x,iszero(f5(x))))

x = find_zero(f5,0,  Order1(),verbose=true)
push!(df,("f5","sieczne",x,iszero(f5(x))))

x = find_zero(f6, 0, verbose=true)
push!(df,("f6","przedzial",x,iszero(f6(x))))

x = find_zero((f6, D(f6)),0, Roots.Newton(),verbose=true)
push!(df,("f6","pochodna",x,iszero(f6(x))))

x = find_zero(f6,0,  Order1(),verbose=true)
push!(df,("f6","sieczne",x,iszero(f6(x))))
```

```julia
DD(f) = x->ForwardDiff.derivative(D(f), float(x))
df2 = DataFrame(Function = String[],Type = String[], X_0 = Float64[], Is_zero = Bool[])

x = find_zero(f2, 0)
push!(df2,("f2","przedzial",x,iszero(f2(x))))

x = find_zero(f2,(-1,1),  FalsePosition())
push!(df2,("f2","falsi",x,iszero(f2(x))))

x = find_zero(f2,(-1,1),  FalsePosition(5),verbose=true)
push!(df2,("f2","falsi5",x,iszero(f2(x))))

x = find_zero((f2, D(f2)),0, Roots.Newton())
push!(df2,("f2","newton",x,iszero(f2(x))))

x = find_zero((f2, D(f2), DD(f2)), 0.0, Roots.Halley())
push!(df2,("f2","halley",x,iszero(f2(x))))

x = find_zero(f2,0,  Order1())
push!(df2,("f2","sieczne",x,iszero(f2(x))))

x = find_zero(f2,0,  Order2())
push!(df2,("f2","steffensen",x,iszero(f2(x))))
```
